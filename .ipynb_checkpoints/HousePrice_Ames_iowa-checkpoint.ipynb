# *Model 1: Logistic Regression*

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
model_log = logreg.fit(xTrain_scale, y)
```

# *Model 3 : Grid Search*

```python
from sklearn.linear_model import SGDClassifier, SGDRegressor

sgd_reg = SGDRegressor(fit_intercept=False, 
                               penalty = None, 
                               max_iter=1000, 
                               alpha = 0.001).fit(xTrain_scale, y)
```

```python
train_preds = grid.predict(xTrain_scale)
test_preds = grid.predict(xTest_scaled)
```

```python
# Check the structure and contents of clf.coef_
print("Structure of clf.coef_:", type(clf.coef_))
print("Contents of clf.coef_:", clf.coef_)

# Verify the shape of clf.coef_
print("Shape of clf.coef_:", clf.coef_.shape)

# If clf.coef_ is an array, check the index obtained from X.columns.get_loc('Overall Qual')
overallqual_index = X.columns.get_loc('Overall Qual')
print("Index of 'Overall Qual':", overallqual_index)
```

```python
#Given the outlook of our data, to target certain columns, we need to take their coeficients, as such I will print out all
#the coeficinets the ridge model has per feature

# Get the names of all features
feature_names = X.columns

# Iterate over each feature and its coefficient
for feature_name, coefficient in zip(feature_names, clf.coef_):
    print("Feature:", feature_name)
    print("Coefficient:", coefficient)
```

```python
# Seeing how our features cause prices to fluctuate

overallqual_index = X.columns.get_loc('Overall Qual')

# Retrieve the coefficient corresponding to the 'Overall Qual' feature
overallqual_coeff = clf.coef_[overallqual_index]

print("Coefficient for 'Overall Qual':", overallqual_coeff)
```

#### *What is the influnce of the overall house quality on the sale price?*

```python
overall_qual_index = 6

# Extract 'Overall Qual' feature from scaled datasets
xTrain_overall_qual = xTrain_scale[:, overall_qual_index]
xTest_overall_qual = xTest_scaled[:, overall_qual_index]

# Apply coefficient to 'Overall Qual' feature
predicted_sale_price_train = xTrain_overall_qual * clf.coef_[overall_qual_index]
predicted_sale_price_test = xTest_overall_qual * clf.coef_[overall_qual_index]

# Visualize the relationship
plt.figure(figsize=(20, 6))
plt.scatter(xTrain_overall_qual, predicted_sale_price_train, color='blue', label='Predicted Sale Price (Train)')
plt.scatter(xTest_overall_qual, predicted_sale_price_test, color='red', label='Predicted Sale Price (Test)')
plt.xlabel('Overall Quality')
plt.ylabel('Predicted Sale Price')
plt.title('Predicted Sale Price vs. Overall Quality')
plt.legend()
plt.grid();
```

What is the influnce of the overall house quality on the sale price?¶
- Answer: An increase in the house quality leads to an equal increase in the sale price.

```python
# Initialize an empty dictionary to store feature indexes
feature_indexes = {}
feature_names = X.columns

# Iterate over each feature and its coefficient
for idx, (feature_name, _) in enumerate(zip(feature_names, clf.coef_)):

    feature_indexes[feature_name] = idx

print("Dictionary of feature indexes:")
print(feature_indexes)
```

#### *What features of a house most impacted its Sale price?*

```python
# Create a list of tuples containing feature name and coefficient
feature_coefficients = list(zip(feature_names, clf.coef_))

# Sort the list based on the absolute magnitude of coefficients
sorted_feature_coefficients = sorted(feature_coefficients, key=lambda x: abs(x[1]), reverse=True)

print("Top features with the highest impact on SalePrice:")
for feature, coefficient in sorted_feature_coefficients:
    print(f"Feature: {feature}, Coefficient: {coefficient}")
```

We can see all the features and the impacts they had on the sale price of a house, the most being Overall Quality and the
least, Low Quality Finished Square Feet

```python
top_n = 10  # Number of top features
bottom_n = 10  # Number of bottom features

# Extract feature names and coefficients
feature_names = X.columns
feature_coefficients = clf.coef_
```

```python
# Sort feature coefficients by their absolute values
sorted_indices = sorted(range(len(feature_coefficients)), key=lambda i: abs(feature_coefficients[i]), reverse=True)
```

```python
# Extract top features and their coefficients
top_features = [feature_names[i] for i in sorted_indices[:top_n]]
top_coefficients = [feature_coefficients[i] for i in sorted_indices[:top_n]]

# Plot top features
plt.figure(figsize=(15, 6))
plt.barh(top_features, top_coefficients, color='blue')
plt.xlabel('Coefficient')
plt.ylabel('Feature')
plt.title('Top Features with Greatest Impact on Sale Price')
plt.gca().invert_yaxis()  # Invert y-axis to display the most impactful feature at the top
plt.grid();
```

As suspected, overall house quality greatly determined the price of a house followed by the Above grade (ground) living area square feet

```python
# Extract bottom features and their coefficients
bottom_features = [feature_names[i] for i in sorted_indices[-bottom_n:]]
bottom_coefficients = [feature_coefficients[i] for i in sorted_indices[-bottom_n:]]
```

```python
# Plot bottom features
plt.figure(figsize=(12, 6))
plt.barh(bottom_features, bottom_coefficients, color='red')
plt.xlabel('Coefficient')
plt.ylabel('Feature')
plt.title('Features with Least Impact on Sale Price')
plt.gca().invert_yaxis()  # Invert y-axis to display the least impactful feature at the top
plt.grid();
```

So from Our machine learning model with an accuracy of of 80%, the features that greatly affected the Sales Price of homes
are the Overall Quality, Above grade (ground) living area square feet, Second floor square feet, and Year the house was built,
