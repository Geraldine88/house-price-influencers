We're are going to;
- 1 - Scale Numerical Features
- 2 - Encode Categorical Features
- 3 - Concatenate the scaled numerical features and encoded categorical features in to a single dataset.

```python
num_feats_train = X_train.select_dtypes(include=['int', 'float']).columns
cat_feat_train = X_train.select_dtypes(include=['object']).columns

#Defining preprocessing steps for numerical and categorical features
num_transform = Pipeline(steps= [('scaler', StandardScaler())])

cat_transform = Pipeline(steps= [('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transform, num_feats_train),
        ('cat', cat_transform, cat_feats_train)
    ])

pipe = Pipeline(steps=[('preprocessor', preprocessor),
                       ('rfe', RFE(estimator=Lasso(alpha=10, max_iter=1000))),
                        ('model', Lasso(alpha = 1000, max_iter=10000))])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
# Evaluate model performance with cross-validation, 5 folds
mse_scores = -cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
print(f"Mean Squared Error:{mse_scores.mean()}")
```
